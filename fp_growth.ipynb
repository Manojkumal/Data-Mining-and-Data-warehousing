```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
from IPython.display import display_html
```

```python
def toy_dataset():
    data = [
        ['Milk','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
        ['Milk', 'Onion','Nutmeg','Kidney Beans','Eggs','Yougurt'],
        ['Milk','Apple','Kidney Beans','Eggs'],
        ['Milk','Unicorn','Corn','Kidney Beans','Eggs'],
        ['Corn','Unicorn','Onion','Kidney Beans','Ice cream','Eggs']
    ]
    print("Raw data:", data)
    te = TransactionEncoder()
    te_ary = te.fit(data).transform(data)
    df = pd.DataFrame(te_ary,columns = te.columns_)
    print("Raw Encoded data:")
    display_html(df)
    return df

data = toy_dataset()
```

```python
def frequent_itemset(data):
    print("Enter the value of minimum support threshold: ")
    support = float(input()) # input = 0.6
    frequent_itemsets = fpgrowth(data, min_support=support, use_colnames=True)
    return frequent_itemsets

frequent_itemsets = frequent_itemset(data)
print(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules

def association_rule(frequent_itemsets):
    print("Enter minimum confidence threshold value: ")
    min_confidence = float(input()) # input 0.75
    rule = association_rules(frequent_itemsets, min_threshold=min_confidence)
    print("The learned association rules: ")
    display_html(rule.drop(['leverage','conviction'],axis=1))
    
# Call the function with appropriate frequent_itemsets input
association_rule(frequent_itemsets)
```
